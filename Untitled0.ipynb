```python
# Paso 1: Importar las librer√≠as necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el CSV en un DataFrame de pandas
mamiferos_acg = pd.read_csv('https://gist.githubusercontent.com/geociencias98/4698f49cc870bfc70415bac5d5e11a6c/raw/cbf979bcd6f4809d03ba40b4d2ffbf6521449909/gistfile1.txt')

# Mostrar las primeras filas para verificar la carga
mamiferos_acg.head()
```

```python
# Filtrar los avistamientos de venados por el valor en 'taxon_id' igual a 42223
venados_acg = mamiferos_acg[mamiferos_acg['taxon_id'] == 42223]

# Contar el n√∫mero de avistamientos por a√±o
avistamientos_venados = venados_acg['year'].value_counts().sort_index()

# Crear un gr√°fico de dispersi√≥n
plt.figure(figsize=(10, 6))
plt.scatter(avistamientos_venados.index, avistamientos_venados.values, color='blue', alpha=0.7)
plt.title('Cantidad de Avistamientos de Venados por A√±o')
plt.xlabel('A√±o')
plt.ylabel('N√∫mero de Avistamientos')
plt.grid(True)

# Mostrar el gr√°fico
plt.show()
```

```python
# Gr√°fico 2: Cantidad de avistamientos de cada una de las 3 especies principales
especies_principales = mamiferos_acg['common_name'].value_counts().nlargest(3)

plt.figure(figsize=(10, 6))
especies_principales.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Avistamientos de las 3 Especies Principales')
plt.xlabel('Especie')
plt.ylabel('N√∫mero de Avistamientos')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtrar los avistamientos de venados por el valor en 'taxon_id' igual a 43355
tapir_acg = mamiferos_acg[mamiferos_acg['taxon_id'] == 43355]

# Contar el n√∫mero de avistamientos por a√±o
avistamientos_tapir = tapir_acg['year'].value_counts().sort_index()

# Crear un gr√°fico de dispersi√≥n
plt.figure(figsize=(10, 6))
plt.scatter(avistamientos_tapir.index, avistamientos_tapir.values, color='green', alpha=0.7)
plt.title('Cantidad de Avistamientos de Tapir por A√±o')
plt.xlabel('A√±o')
plt.ylabel('N√∫mero de Avistamientos')
plt.grid(True)

# Mostrar el gr√°fico
plt.show()
```
